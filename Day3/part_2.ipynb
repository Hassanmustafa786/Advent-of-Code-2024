```python
# Day 3 (Part 2)

import re

# Define file path
file_path = "Part2.txt"

# Read the file
with open(file_path, 'r') as file:
    content = file.read()

print(content)
```

```python
mul_matches = re.findall(r"(mul\((\d{1,3}),(\d{1,3})\)|do\(\)|don't\(\))", content)
print("Mul Matches Length:", len(mul_matches), "\n", mul_matches)
```

```python
import re

# Results list
results = []
allow_multiplication = True  # Flag to control whether `mul()` results are added

for expr in mul_matches:
    expression = expr[0]  # Unpack the tuple
    print("Expression:", expression)

    # Match for `mul(num1, num2)`
    mul_match = re.match(r"mul\((\d+),(\d+)\)", expression)
    if mul_match:
        num1, num2 = map(int, mul_match.groups())  # Extract and convert to integers
        if allow_multiplication:
            result = num1 * num2
            results.append(result)
            print(f"Multiplying {num1} and {num2} gives {result}")
        else:
            print(f"Skipping multiplication for {expression} due to 'don't()' state.")
        continue

    # Match for `do()`
    do_match = re.match(r"do\(\)", expression)
    if do_match:
        allow_multiplication = True  # Allow `mul()` results to be added again
        results.append("do()")
        print(f"Matched: {do_match.group()}")
        continue

    # Match for `don't()`
    dont_match = re.match(r"don't\(\)", expression)
    if dont_match:
        allow_multiplication = False  # Disable adding `mul()` results
        results.append("don't()")
        print(f"Matched: {dont_match.group()}")
        continue

    # If no match
    print("No match found for:", expression)

# Print the results
print("Final Results Length:", len(results))
print("Final Results:", results)
```

```python
# Use a separate list to store valid elements
filtered_results = []

for i in results:
    if i not in ["don't()", "do()"]:  # Exclude "don't()" and "do()"
        filtered_results.append(i)

# Replace the original list with the filtered one
results = filtered_results

print("Updated Results:", results)
```

```python
total_sum = sum(results)
print(total_sum)
```
