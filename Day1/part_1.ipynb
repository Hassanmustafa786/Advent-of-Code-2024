```python
# Day 1 (Part 1)
# Define file path
file_path = "Part1.txt"

# Initialize two empty lists
first_column = []
second_column = []

# Read the file
with open(file_path, 'r') as file:
    for line in file:
        # Split the line into two parts
        columns = line.split()
        # Append the values to respective lists
        first_column.append(int(columns[0]))
        second_column.append(int(columns[1]))

# Print the results
print("First Column:", first_column)
print("Second Column:", second_column)
```

```python
# Sort the lists
first_column_sorted = sorted(first_column)
second_column_sorted = sorted(second_column)

# Print the sorted results
print("Sorted First Column:", first_column_sorted)
print("Sorted Second Column:", second_column_sorted)
```

```python
# Compute the difference
difference_column = [second - first for first, second in zip(first_column_sorted, second_column_sorted)]

# Print the sorted lists and the difference
print("Sorted First Column:", first_column_sorted)
print("Sorted Second Column:", second_column_sorted)
print("Difference Column:", difference_column)
```

```python
difference_column_abs = []
for i in difference_column:
    difference_column_abs.append(abs(i))
    
print("Absolute Difference Column:", difference_column_abs)
```

```python
# Calculate the total sum of the difference column
total_difference_sum = sum(difference_column_abs)

# Print the total sum
print("Total Sum of Difference Column:", total_difference_sum)
```
